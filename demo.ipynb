```python
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score
from preprocessing_data import prepare_data, train_data, test_data
from softmax_regression.soft_reg import MySoftmaxRegression
from logistic_regression.log_reg import MyLogisticRegression
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
# My Logistic Regression Model

train_images, train_labels, test_images, test_labels = prepare_data(train_data=train_data, test_data=test_data, digits=[0, 1])

# Train Model
logistic_model = MyLogisticRegression()
logistic_model.fit(train_images, train_labels)

# Predict
y_pred = logistic_model.predict(test_images)

# Score
print("Accuracy:", accuracy_score(test_labels, y_pred))
print("Precision:", precision_score(test_labels, y_pred))
print("Recall:", recall_score(test_labels, y_pred))
print("F1 Score:", f1_score(test_labels, y_pred))
```

```python
# My Softmax Regression Model

train_images, train_labels, test_images, test_labels = prepare_data(train_data=train_data, test_data=test_data,digits=range(10))

# Train Model
softmax_model = MySoftmaxRegression(n_classes=10)
softmax_model.fit(train_images, train_labels)

# Predict
y_pred = softmax_model.predict(test_images)

# Score
print("Accuracy:", accuracy_score(test_labels, y_pred))
print("Precision:", precision_score(test_labels, y_pred, average="macro"))
print("Recall:", recall_score(test_labels, y_pred, average="macro"))
print("F1 Score:", f1_score(test_labels, y_pred, average="macro"))
```
